```python
#Average Rate of Change to Instantaneous Rate of Change
def average_rate_of_change(function, a, b):
    return (function(b) - function(a)) / (b - a)
```

```python
def distance(t):
    return 4.9 * t**2
```

```python
#a
avg1 = average_rate_of_change(distance, 5, 6)

print("Average speed from 5 to 6 seconds:", avg1)
```

```python
#b
avg2 = average_rate_of_change(distance, 5, 5.5)

print("Average speed from 5 to 5.5 seconds:", avg2)
```

```python
#c
avg3 = average_rate_of_change(distance, 5, 5.1)

print("Average speed from 5 to 5.1 seconds:", avg3)
```

```python
#d
intervals = [5.5, 5.1, 5.01, 5.001]

for t in intervals:
    avg = average_rate_of_change(distance, 5, t)
    print(f"Average speed from 5 to {t}: {avg}")
```

#d ans= As the speed gets smaller, the average speed is 49 m/s.

```python
#e: find the deriative
import sympy as sp

# Define the variable
t = sp.symbols('t')

# Define the function f(t) = 4.9 * t^2
f = 4.9 * t**2

# Differentiate f with respect to t
f_prime = sp.diff(f, t)

print("f(t) =", f)
print("f'(t) =", f_prime)
```

```python
#f: Evaluate the deriative
instantaneous_speed = 9.8 * 5

print("Instantaneous speed at t = 5:", instantaneous_speed)
```

#f written portion answer: The derivative gives an instantaneous speed of 49 m/s, which matches the value in part d. As the time interval becomes smaller, the average rate of change approaches the derivative, which represents the instantaneous rate of change.

```python
#Calculating and Interpreting Partial Derivatives
def price(C, Y):
    return 16000 + 2400*C - 1800*Y

selling_price = price(8, 5)

print("Predicted Selling Price: $", selling_price)
```

```python
#b Find and interpret value
partial_C = 2400

print("∂P/∂C =", partial_C)
```

#b interpretation: Each condition point adds $2,400 to the price, assuming the car's age is the same. Better condition means a higher price.

```python
#c Find and interpret value 
partial_Y = -1800

print("∂P/∂Y =", partial_Y)
```

#c interpretation: For each year the car ages, the predicted selling price decreases by $1,800, assuming the condition rating remains constant.
