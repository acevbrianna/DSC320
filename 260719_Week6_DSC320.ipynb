```python
#Import libraries
import pandas as pd
import numpy as np
```

```python
# 1 Create normalization
def normalize(data):
    """
    Normalizes a vector so all values are between 0 and 1.
    """

    data = np.array(data)

    normalized = (data - np.min(data)) / (np.max(data) - np.min(data))

    return normalized
```

```python
#Test
sample = [5, 10, 15, 20, 25]

print(normalize(sample))
```

```python
# 2 Create standardization function
def standardize(data):
    """
    Standardizes a vector using the z-score formula.
    """

    data = np.array(data)

    standardized = (data - np.mean(data)) / np.std(data, ddof=1)

    return standardized
```

```python
#Test
sample = [5, 10, 15, 20, 25]

print(standardize(sample))
```

```python
#3D 
housing_features = housing[
    [
        "housing_median_age",
        "median_income",
        "bedrooms_per_household"
    ]
]

housing_features.head()
```

```python
#3E
#Apply standardization to each feature
standardized_df = housing_features.copy()

for column in standardized_df.columns:
    standardized_df[column] = standardize(
        standardized_df[column]
    )

standardized_df.head()
```
