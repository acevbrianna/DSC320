```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define Function
# Function: f(x) = sqrt(4 - x^2)

def f(x):
    return np.sqrt(4 - x**2)
```

```python
#Create the function
def right_riemann_sum(func, a, b, n):
    """
    Calculates the Right Riemann Sum of a function.

    Parameters:
        func : mathematical function
        a    : left endpoint
        b    : right endpoint
        n    : number of rectangles

    Returns:
        Right Riemann Sum
    """

    dx = (b - a) / n
    total = 0

    for k in range(1, n + 1):
        x = a + k * dx
        total += func(x)

    return total * dx
```

```python
#B Calculate R4
R4 = right_riemann_sum(f, 0, 2, 4)

print("Right Riemann Sum (R4):", R4)
```

```python
#Plot the function
x = np.linspace(0, 2, 400)
y = f(x)

plt.figure(figsize=(7,5))
plt.plot(x, y, label='f(x)=sqrt(4-x²)', linewidth=2)

plt.title("f(x)=√(4−x²)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.legend()
plt.show()
```

```python
#Plot the rectangles
n = 4
a = 0
b = 2

dx = (b - a) / n

x = np.linspace(a, b, 400)
plt.figure(figsize=(7,5))
plt.plot(x, f(x), color='blue', linewidth=2)

for k in range(1, n + 1):
    right = a + k * dx
    left = right - dx

    plt.bar(left,
            f(right),
            width=dx,
            align='edge',
            alpha=0.4,
            edgecolor='black')

plt.title("Right Riemann Sum (R4)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```

```python
#C Approximate Area using larger values of n
values = [10, 50, 100, 500, 1000, 5000]

print("Right Riemann Sum Approximations\n")

for n in values:
    area = right_riemann_sum(f, 0, 2, n)
    print(f"n = {n:<5} Area ≈ {area:.8f}")
```

```python
print("As the number of rectangles increases,")
print("the Right Riemann Sum approaches approximately 3.14159.")
print()
print("This value is π (pi).")
print("The graph y = sqrt(4 - x^2) is the upper half of a circle")
print("with radius 2.")
print("Therefore, the area under the curve from x = 0 to x = 2")
print("is one-quarter of the area of a circle with radius 2:")
print()
print("Area = (1/4) * π * (2^2) = π")
```
